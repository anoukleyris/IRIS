```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

train_data, test_data = train_test_split(data, test_size=0.2, random_state=42) 
train_data, val_data = train_test_split(train_data, test_size=0.25, random_state=42)

feature = ['sepal_length', 'petal_length', 'petal_width','sepal_width']
target = 'species' 


X_train = train_data[feature]
y_train = train_data[target]

X_val = val_data[feature]
y_val = val_data[target]
X_test = test_data[feature]
y_test = test_data[target]

# Define the model
model = LogisticRegression()
model.fit(X_train, y_train)

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10]}

grid_search = GridSearchCV(model, param_grid=param_grid, cv=5)

grid_search.fit(X_train, y_train)

print("Best hyperparameters: ", grid_search.best_params_)

y_test_pred = grid_search.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Test accuracy: ", test_accuracy)


def Survived(model,sepal_length, sepal_width, petal_length, petal_width):
    x = np.array([sepal_length, sepal_width, petal_length, petal_width]).reshape(1, 4)
    prediction = model.predict(x)
    probability = model.predict_proba(x)
    print("Prediction:", prediction)
    print("Probabilities:", probability)

Survived(model, 4,3,1.3,0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import learning_curve

dt = DecisionTreeClassifier()
train_sizes, train_scores, test_scores = learning_curve(
    dt, X_train, y_train, cv=5, n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='accuracy')
plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training score')
plt.plot(train_sizes, np.mean(test_scores, axis=1), 'o-', label='Validation score')
plt.xlabel('Training examples')
plt.ylabel('Score')
plt.legend(loc='best')
plt.title('Evolves of the accuracy test as a function of the number of samples')
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import load_iris
from sklearn.svm import SVC


iris = load_iris()
X = iris.data[:, [0, 2]]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)#comme avant, 30% pour le test

clf = SVC(kernel='linear', C=1)

clf.fit(X_train, y_train)


plot_decision_regions(X_test, y_test, clf=clf, legend=2)

plt.xlabel('sepal length (cm)')
plt.ylabel('petal length (cm)')
plt.title('Decision boundaries SVM on Iris')


plt.show()

#les formes c'est pour chaque plante de chaque espÃ¨ce et on a quaznd mÃªme une trÃ¨s bonne distinction entre les trois espÃ¨ces sauf frontiÃ¨re vert orange 
```
